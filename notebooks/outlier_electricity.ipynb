```python
import pandas as pd 
df=pd.read_csv("../Data/final/environment_data.csv")
```

```python
df["Year"]=pd.to_datetime(df["Date"]).dt.year
df["Month"]=pd.to_datetime(df["Date"]).dt.month
monthly_electricity = df.groupby(
    ['City', 'Year', 'Month'],
    as_index=False
)['Electricity Consumption'].mean()
```

```python
def detect_electricity_outliers(group):

    q1 = group['Electricity Consumption'].quantile(0.25)
    q3 = group['Electricity Consumption'].quantile(0.75)

    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    group['Outlier'] = (
        (group['Electricity Consumption'] < lower) |
        (group['Electricity Consumption'] > upper)
    )

    return group
```

```python
monthly_outliers = monthly_electricity.groupby(
    'City',
    group_keys=False
).apply(detect_electricity_outliers)
```
