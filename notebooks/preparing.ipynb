##merge air quality,greenspace

##handle global_weather take from it nairobi,london


```python
df5=pd.read_csv("../Data/raw/global_weather_100_cities_2014_2024.csv")
```

```python
df5['date'] = pd.to_datetime(df5['date'])
```

```python
df_2023 = df5[df5['date'].dt.year == 2023]
```

```python
df_2023["city"].value_counts()
```

```python
cities = ['London', 'Nairobi']
df_filtered = df_2023[df_2023['city'].isin(cities)]
```

```python
df_filtered["city"].value_counts()
```

```python
df_filtered.columns
```

```python
df_filtered = df_filtered.rename(columns={
    'avg_temperature_c': 'Temperature',
    'avg_humidity_pct': 'humidity',
    'city': 'City',
    "max_wind_speed_kmh":"wind_speed"
})
```

```python
df_filtered = df_filtered.drop(columns=['rainfall_mm'])
```

```python
df_filtered.sample(10)
```

```python


df_filtered.to_csv('../Data/raw/weather_Nairobi_london.csv', index=False)
```
